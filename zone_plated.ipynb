```python
import numpy as np
import cv2
import math
import matplotlib.pyplot as plt
```

```python
N = 501
x2 = (N-1)/2

x1 = -x2

x, y = np.mgrid[x1:x2+1, x1:x2+1]
r = np.hypot(x,y)

km = 0.7*math.pi
rm = x2
w = rm/10
term1 = np.sin((km*(r**2)/(2*rm)))
term2 = 0.5*np.tanh((rm - r)/w) + 0.5
g = term1 * term2
I = (g + 1)/2
```

```python
img = I*256
img = img.astype("int")
plt.imshow(img, cmap = 'gray')
```

```python
I_cropped = img[125:375, 125:375]
print(I_cropped.shape)
plt.imshow(I_cropped, cmap = 'gray')
```
