```python
# q2
import cv2
import numpy as np
from matplotlib import pyplot as plt
img1 = plt.imread('Fig1.png')
img = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
plt.imshow(img,cmap='gray')
```

```python
cross=cv2.getStructuringElement(cv2.MORPH_CROSS,(22,22))
im1_hf = cv2.morphologyEx(img, cv2.MORPH_CROSS, cross)
plt.figure(figsize=(16,16))
plt.subplot(1,2,1)
plt.imshow(img,cmap='gray')
plt.title('Original image')
plt.subplot(1,2,2)
plt.imshow(im1_hf,cmap = 'gray')
plt.title('Hole-filled image')
```

```python
#Difference of Gaussian: Extracting boundary
I_gaussian_blur1 = cv2.GaussianBlur(img,(5,5),3,cv2.BORDER_DEFAULT)
I_gaussian_blur2 = cv2.GaussianBlur(img,(5,5),7,cv2.BORDER_DEFAULT)
DOG=I_gaussian_blur1-I_gaussian_blur2
plt.imshow(DOG,cmap='gray')
plt.title('Difference of Gaussian')
plt.show()
```

```python
# q2
img = plt.imread('Fig2.tif')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img,cmap = 'gray')
```

```python
# q3
se = cv2.getStructuringElement(cv2.MORPH_RECT,(5,5))
im1_dilate = cv2.dilate(img, se, iterations=1)
connectivity=4
output=cv2.connectedComponentsWithStats(im1_dilate,connectivity)
num_labels=output[0]
labels=output[1]
stats=output[2]
centroids=output[3]
print(num_labels)
```

```python
# q3
img = plt.imread('download.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img,cmap = 'gray')
```

```python
img = 255 * img
img1 = img.astype(np.uint8)
plt.imshow(img1,cmap = 'gray')
```

```python
y=0
x=0
h=250
w=250
Icropped= img1[y:y+h, x:x+w]
plt.imshow(Icropped,cmap = 'gray')
```

```python
kernel1 = np.ones((5,5),np.float32)/25
I_gaussian_blur1 = cv2.GaussianBlur(Icropped,(5,5),2,cv2.BORDER_DEFAULT)
plt.figure(figsize=(10,10))

plt.imshow(I_gaussian_blur1,cmap = 'gray')
plt.title('smoothing with Gaussian sigma=2')
```

```python
I_gaussian_blur1 = cv2.GaussianBlur(img,(5,5),3,cv2.BORDER_DEFAULT)
I_gaussian_blur2 = cv2.GaussianBlur(img,(5,5),7,cv2.BORDER_DEFAULT)
DOG=I_gaussian_blur1-I_gaussian_blur2
plt.imshow(DOG,cmap='gray')
plt.title('Difference of Gaussian')
plt.show()
```
